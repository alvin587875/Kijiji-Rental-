```python
# Import pandas for data manipulation and analysis.
import pandas as pd
import numpy as np
df_3 = pd.read_csv("CSDpop.csv")
df_2 = pd.read_csv("kijiji_dataset_01.csv")
```

```python
df_2.head(1)
```

```python
df_2 = df_2.drop(["adId","Title","Location","URL","Date Posted","CSDUID"], axis = 1) # Droping unnecessary columns
```

```python
df_3.head()
```

```python
obj_df=df_2.copy()
```

```python
str_num_conv = {"Bedrooms":{"Bedrooms: 2": 2,"Bedrooms: 1": 1,"Bedrooms: 3": 3,"Bedrooms: 4": 4 ,      # Encoding Bedrooms columns
                            "Bedrooms: 1 + Den": 1.5 ,"Bedrooms: Bachelor/Studio": 0.5 ,
                            "Bedrooms: 2 + Den": 2.5 ,"Bedrooms: 3 + Den": 3.5 ,"Bedrooms: 4 + Den": 4.5 ,"Bedrooms: 5+": 5},
  }
```

```python
obj_df = obj_df.replace(str_num_conv)
obj_df['Bedrooms']=obj_df['Bedrooms'].astype(float)
obj_df.head()
```

```python
str_num_conv = {"Bathrooms":{"Bathrooms: 1": 1,"Bathrooms: 2": 2,"Bathrooms: 2.5": 2.5,"Bathrooms: 1.5": 1.5 ,  # Encoding Bathrooms columns
                            "Bathrooms: 3": 3 ,"Bathrooms: 3.5": 3.5 ,
                            "Bathrooms: 4": 4 ,"Bathrooms: 5": 5 ,"Bathrooms: 4.5": 4.5 ,"Bathrooms: 5.5": 5.5, "Bathrooms: 6+":6},
  }
```

```python
obj_df = obj_df.replace(str_num_conv)
obj_df['Bedrooms']=obj_df['Bedrooms'].astype(float)
obj_df.head()
```

```python
str_num_conv = {"Hydro":{"N": 0,"Y": 1}} # Encoding Hydro columns
obj_df = obj_df.replace(str_num_conv)
obj_df['Hydro']=obj_df['Hydro'].astype(object)

obj_df.head()
```

```python
str_num_conv = {"Heat":{"N": 0,"Y": 1}} # Encoding Heat columns
obj_df = obj_df.replace(str_num_conv)
obj_df['Heat']=obj_df['Heat'].astype(object)

obj_df.head()
```

```python
str_num_conv = {"Water":{"N": 0,"Y": 1}} # Encoding Water columns
obj_df = obj_df.replace(str_num_conv)
obj_df['Water']=obj_df['Water'].astype(object)
obj_df.head()
```

```python
obj_df['Price'] = obj_df['Price'].str.replace('$', '')   # Removing the special characteristics
obj_df['Price'] = obj_df['Price'].str.replace(',', '')
obj_df['Price'] = pd.to_numeric(obj_df['Price'], errors='coerce')
obj_df['Price']=obj_df['Price'].astype(float)
obj_df.head()
```

```python
str_num_conv = {"Heat":{"N": 0,"Y": 1}}
obj_df = obj_df.replace(str_num_conv)
obj_df['Heat']=obj_df['Heat'].astype(float)
```

```python
str_num_conv = {"Water":{"N": 0,"Y": 1}}
obj_df = obj_df.replace(str_num_conv)
obj_df['Water']=obj_df['Water'].astype(float)
```

```python
str_num_conv = {"Agreement Type":{"1 Year": 2,'Month-to-month': 1,'Not Available':0}}       # Encoding Agreement type columns
obj_df = obj_df.replace(str_num_conv)
obj_df['Agreement Type']=obj_df['Agreement Type'].astype(float)
```

```python
obj_df["Size"] = pd.to_numeric(obj_df["Size"], errors='coerce')
```

```python
#obj_df["Size"].fillna(obj_df["Size"].mean,inplace= True)              # replacing the empty vlaue in size with mean value
mean_size = obj_df["Size"].mean()
obj_df['Size'].replace(to_replace=pd.NA, value=mean_size, inplace=True)
```

```python
obj_df = obj_df[obj_df["Size"] > 300]  # removing outliear, we defined upper 9000 and lower 300 boundary
```
